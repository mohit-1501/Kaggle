```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
dfa=pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/shops.csv")
dfb=pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/item_categories.csv")
dfc=pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/test.csv")
dfd=pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv")
dfe=pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/items.csv")
```

```python
df=pd.merge(dfd,dfe)
df=pd.merge(df,dfb);
df=pd.merge(df,dfa);
df
```

```python
df=df.drop(["date"],axis=1)
df=df.drop(["item_name","item_category_name",'shop_name'],axis=1)
ll=df.groupby([c for c in df.columns if c not in['item_cnt_day']],as_index=False)["item_cnt_day"].sum()
ll=ll.rename(columns={"item_cnt_day":"item_cnt_month"})
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
import xgboost as xgb
from sklearn import metrics
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
from xgboost import XGBRegressor
X = ll[ ['shop_id',"item_id"]].values
y = ll['item_cnt_month'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
xf=XGBRegressor(objective='reg:squarederror')
xf.fit(X_train,y_train)
```

```python
xpred=dfc[["shop_id","item_id"]].values
Main_pred=xf.predict(xpred)
```

```python
ab=dfc
ab["item_cnt_month"]= Main_pred
ab=ab.drop(["shop_id","item_id"],axis=1)
ab
```

```python
ab.to_csv("E:\\xcd.csv",index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

# model
model = RandomForestRegressor(n_estimators=50, random_state=0, n_jobs=-1, max_features= 'log2')
   
    
# train
model.fit(X_train,  y_train)

#score
print('Train score is : ' , model.score(X_train, y_train))
print('Test  score is : ' , model.score(X_test , y_test ))
```

```python
Main_pred2=model.predict(xpred)
aa=dfc
aa["item_cnt_month"]= Main_pred2
aa=aa.drop(["shop_id","item_id"],axis=1)
aa.to_csv("E:\\axcd.csv",index=False)
```
