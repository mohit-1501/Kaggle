```python
import matplotlib.pyplot as plt
import numpy as np
f, ax = plt.subplots(1)
xdata = [0, 20,25]
ydata = [10, 30,35]
x = np.arange(60)
y = np.arange(60)
plt.grid(False)
plt.xticks(x, " ")
plt.yticks(y, " ")
plt.xlabel("Time")
plt.ylabel("Accumulators Activity")
by=[10,10,10]
ty=[30,30,30]
ax.plot(xdata,by,label='BASELINE')
ax.plot(xdata,ty,label='THRESHOLD')
ax.plot(xdata, ydata)
ax.set_ylim(ymin=0,ymax=40)
ax.set_xlim(xmin=0,xmax=25)
plt.legend()
plt.show(f)
```
