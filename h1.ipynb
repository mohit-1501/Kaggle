```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math as m
```

```python
#OLD EXPERIMENT
x=[]
y=[]
for i in range (10, 150):
    x.append(i)
    z=m.log(i)
    y.append(z)
    
#MODIFIED EXPERIMENT  
p=[]
q=[]
for j in range(5,150):
    p.append(j)
    z=1.05*(m.log(j))           #INCREASED RATE OF DILATION BY 0.05 FACTOR
    q.append(z)
    
```

```python
plt.plot(x,y)
plt.plot(p,q)
plt.plot([7,7],[1.5,5], 'k-', lw=1,dashes=[2, 2])
plt.xticks([])
plt.xlabel("Time elapsed")
plt.ylabel("Dilation amount")
plt.yticks([])
plt.title("Dilation of eyes with time")
plt.legend(["Original","Modified","Normal Dilation Reference"])
plt.show()

```
